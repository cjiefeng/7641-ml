```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
model_colors = {
    'DT': '#86C7F3',      # Blue
    'Boosted DT': '#FFD699',  # Orange
    'MLP': '#B4EB8E',     # Green
    'SVM': '#FFA8A8',     # Red
    'KNN': '#D8B4FF'      # Purple
}
```

```python
income_scores = {
  'DT': 0.6689678978253366,
  'Boosted DT': 0.7096107475025836,
  'MLP': 0.6893172165958837,
  'SVM': 0.6671630677587491,
  'KNN': 0.6252170892671066
}
sorted_income_scores = dict(sorted(income_scores.items(), key=lambda item: item[1]))

# Plotting
plt.barh(list(sorted_income_scores.keys()), list(sorted_income_scores.values()), color=[model_colors[model] for model in sorted_income_scores.keys()])
plt.xlabel('Income Scores')
plt.title('Income Scores for Different ML Models')
plt.grid(axis='x')

# Show plot
plt.show()
```

```python
bank_scores = {
  'DT': 0.6244875860468714,
  'Boosted DT': 0.6404170881643452,
  'MLP': 0.6155957378107846,
  'SVM': 0.6228406909788867,
  'KNN': 0.5832668050455053
}
sorted_bank_scores = dict(sorted(bank_scores.items(), key=lambda item: item[1]))

# Plotting
plt.barh(list(sorted_bank_scores.keys()), list(sorted_bank_scores.values()), color=[model_colors[model] for model in sorted_bank_scores.keys()])
plt.xlabel('Bank Scores')
plt.title('Bank Scores for Different ML Models')
plt.grid(axis='x')

# Show plot
plt.show()
```

```python
# Sample data
categories = ['DT', 'Boosted DT', 'MLP', 'SVM', 'KNN']
score1 = [138, 705, 5850, 17500, 23]
score2 = [2, 12, 17, 1710, 1560]

# Create indices for the categories
ind = np.arange(len(categories))

# Width of each bar
width = 0.35

# Create horizontal bar chart with two scores for each category
plt.barh(ind, score1, width, label='Training Time')
plt.barh(ind + width, score2, width, label='Query Time')

plt.xlabel('Time (ms)')
plt.title('Time by Model')
plt.yticks(ind + width / 2, categories)
plt.xscale('log') 
plt.legend(loc='best')
plt.show()
```
