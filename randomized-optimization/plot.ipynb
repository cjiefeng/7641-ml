```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = {
  'algorithm': ['adam', 'rhc', 'sa', 'ga'],
  'f1': [0.6795131845841785, 0.24566737610215872, 0.4506424318395487, 0.5572492024104927]
}

df = pd.DataFrame(data)

# Plot bar chart
plt.figure(figsize=(8, 6))
plt.bar(df['algorithm'], df['f1'], color=plt.rcParams['axes.prop_cycle'].by_key()['color'])
plt.xlabel('Algorithm')
plt.ylabel('F1 Score')
plt.title('F1 Score by Algorithm')
plt.ylim(0, 1)  # Set y-axis limit to ensure proper visualization of scores
plt.grid()
plt.show()
```

```python
data = {
  'algorithm': ['adam', 'rhc', 'sa', 'ga'],
  'time': [22.5, 216.9, 84.5, 6151]
}

df = pd.DataFrame(data)

# Plot bar chart
plt.figure(figsize=(8, 6))
plt.bar(df['algorithm'], df['time'], color=plt.rcParams['axes.prop_cycle'].by_key()['color'])
plt.xlabel('Algorithm')
plt.ylabel('Time (s)')
plt.title('Search time by Algorithm')
plt.yscale('log') 
plt.grid()
plt.show()
```
