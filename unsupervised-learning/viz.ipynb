```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
f1_score_mat = np.array([
  [0.6966680378445085, 0.6893172165958837], 
  [0.6624764299182904, 0.6503719447396387]])

class_labels_x = ['Train f1', 'Test f1']
class_labels_y = ['Original', 'PCA']

sns.heatmap(
  f1_score_mat, 
  annot=True, 
  xticklabels=class_labels_x, 
  yticklabels=class_labels_y,
)
plt.title('f1 score heatmap')
plt.show()
```
